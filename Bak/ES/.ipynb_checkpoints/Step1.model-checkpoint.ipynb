```python
import os, random, time
import xgboost
import datetime
import pygam
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from fbprophet import Prophet
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import r2_score
```

```python
y_inven = pd.read_csv('./data/Y_Inven.csv')
x_sales = pd.read_csv('./data/X_Sales.csv')
x_product = pd.read_csv('./data/X_Product.csv')
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor as gpr
from sklearn.gaussian_process.kernels import WhiteKernel, ConstantKernel as C, RBF
```

```python
def dataset(x_dta, case=1, is_train=True):
    x_prev_col = [s for s in x_dta.columns.values if 'Prev' in s]
    x_post_col = [s for s in x_dta.columns.values if 'Post' in s]
    x_var_col = [s for s in x_dta.columns.values if 'Var' in s]

    X_prev_ = x_dta[x_prev_col]
    X_post_ = x_dta[x_post_col]
    X_var_ = x_dta[x_var_col]
    if case ==1 :
        Y_ =  x_dta['Sales']
    else :
        Y_ =  x_dta['Products']

    if is_train :
        X_prev_train = X_prev_[x_dta['YEAR']==2020]
        X_post_train = X_post_[x_dta['YEAR']==2020]
        X_var_train = X_var_[x_dta['YEAR']==2020]
        Y_train = Y_[x_dta['YEAR']==2020]
    else:
        X_prev_train = X_prev_[x_dta['YEAR']==2021]
        X_post_train = X_post_[x_dta['YEAR']==2021]
        X_var_train = X_var_[x_dta['YEAR']==2021]
        Y_train = Y_[x_dta['YEAR']==2021]
    
    return Y_train, X_prev_train, X_post_train, X_var_train    
```

```python
model_sales=xgboost.XGBRegressor(n_estimators=50, learning_rate=0.08, gamma=0, subsample=0.55, colsample_bytree=1, max_depth=5)#, tree_method='gpu_hist', gpu_id=0)
model_products=xgboost.XGBRegressor(n_estimators=50, learning_rate=0.08, gamma=0, subsample=0.55, colsample_bytree=1, max_depth=5)#, tree_method='gpu_hist', gpu_id=0)
```

```python
Y_train, prev_X_train, post_X_train, var_X_train = dataset(x_sales, 1)

model_sales.fit(var_X_train, Y_train)
sales_prev = model_sales.predict(var_X_train)

Y_test, _, _, var_X_test = dataset(x_sales, 1, False)
sales_hat = model_sales.predict(var_X_test)
Y_sales_hat = np.concatenate((sales_prev, sales_hat))
Y_sales = np.concatenate((Y_train, Y_test))
```

```python
Y_train, prev_X_train, post_X_train, var_X_train = dataset(x_product, 2)

model_products.fit(var_X_train, Y_train)
products_prev = model_products.predict(var_X_train)

Y_test, _, _, var_X_test = dataset(x_product, 2, False)
products_hat = model_products.predict(var_X_test)
Y_products_hat = np.concatenate((products_prev, products_hat))
Y_products = np.concatenate((Y_train, Y_test))
```

```python
plt.figure(figsize=(30,20))
plt.subplot(2,1,1)
plt.title("Sales Trend & Prediction")
plt.plot(Y_sales, color='red', marker='o')
plt.plot(Y_sales_hat, color='blue', linestyle='dashed', marker='.')
plt.axvline(43, c='k')
plt.subplot(2,1,2)
plt.title("Products Trend & Prediction")
plt.plot(Y_products, color='red', marker='o')
plt.plot(Y_products_hat, color='blue', linestyle='dashed', marker='.')
plt.axvline(43, c='k')
```

```python
Y_train, prev_X_train, post_X_train, var_X_train = dataset(x_product,2 )
X_train = pd.concat([var_X_train, post_X_train['Post_Product_1']], axis=1)
model_products.fit(X_train, Y_train)
products_prev = model_products.predict(X_train)

Y_test, _, post_X_test, var_X_test = dataset(x_product, 2, False)
X_test = pd.concat([var_X_test, post_X_test['Post_Product_1']], axis=1)

products_hat = model_products.predict(X_test)
Y_products_hat = np.concatenate((products_prev, products_hat))
Y_products = np.concatenate((Y_train, Y_test))
```

```python
Y_sales_train, _, sales_post_X_train, _ = dataset(x_sales, 1)
Y_sales_test, _, sales_post_X_test, _ = dataset(x_sales, 1, False)
Y_products_train, _, products_post_X_train, _ = dataset(x_product, 2)
Y_products_test, _, products_post_X_test, _ = dataset(x_product, 2, False)
y_inven_train = y_inven[y_inven['YEAR']==2020]
y_inven_test = y_inven[y_inven['YEAR']==2021]
```

```python
dta_train = y_inven_train.iloc[2:,:].copy()
dta_train['Sales'] = Y_sales_train.iloc[2:]
dta_train['Product'] = Y_products_train.iloc[2:]
dta_train = dta_train.reset_index(drop=True)
dta_train['prev_Product'] = Y_products_train.iloc[:42]
#dta_train['prev_INVEN'] = np.array(y_inven_train['INVENTORY_PACK_D'].iloc[1:43])
```

```python
dta_test = y_inven_test.copy()
dta_test['Sales'] = Y_sales_test
dta_test['Product'] = Y_products_test
dta_test = dta_test.reset_index(drop=True)
dta_test['prev_Product'] = np.concatenate((np.array(Y_products_train.iloc[42:]), np.array(Y_products_test.iloc[:11])))
#dta_test['prev_INVEN']= np.concatenate((np.array(y_inven_train['INVENTORY_PACK_D'].iloc[43:]), np.array(y_inven_test['INVENTORY_PACK_D'].iloc[:12])))
```

```python
dta_train1 = dta_train.loc[7:].copy().reset_index(drop=True)
dta_test = dta_test.reset_index(drop=True)
#print(dta_train1.columns, '\n',dta_test.columns)
```

```python
train_col_name = list(['Sales', 'Product','prev_Product'])#+prev_inven_col[:1]
model_inven=xgboost.XGBRegressor(n_estimators=50, learning_rate=0.08, gamma=0, subsample=0.7, colsample_bytree=1, max_depth=7)
model_inven.fit(dta_train1[train_col_name], dta_train1['INVENTORY'])
inven_prev = model_inven.predict(dta_train1[train_col_name])

inven_hat = model_inven.predict(dta_test[train_col_name])
Y_inven_hat = np.concatenate((inven_prev, inven_hat))
Y_inven = np.concatenate((dta_train1['INVENTORY'], dta_test['INVENTORY']))
Y_m1_hat = Y_inven_hat
Y_m1 = Y_inven
```

```python
fcst_avg = np.mean(1- np.abs(Y_inven[35:] - Y_inven_hat[35:]) / Y_inven[35:]) * 100
real_avg = np.mean(1- np.abs(Y_inven[:35] - Y_inven_hat[:35]) / Y_inven[:35]) * 100

print(" Real Inventory Mean Average : %f5"%(real_avg))
print(" Forecst Inventory Mean Average : %f5"%(fcst_avg))

plt.figure(figsize=(30,10))
plt.title("Inventory Trend & Prediction")
plt.plot(Y_inven, color='blue', marker='o')
plt.plot(Y_inven_hat, color='red', linestyle='dashed', marker='o')
plt.axvline(34, c='k')
```
